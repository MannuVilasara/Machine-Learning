```python
x = df.iloc[:, 0:2]
y = df.iloc[:, -1]
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
x_train.info()
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(x_train, y_train)
```

```python
y_pred = model.predict(x_test)
```

```python
df.head(4)
```

```python
b0 = model.intercept_
b = model.coef_
b1 = b0
y = (b[0] *1.05	) + (b[1] * 253) + b0
```

```python
y
```

```python
import plotly.graph_objects as go
x_values = np.linspace(
    df['funding'].min(),
    df['funding'].max(),
    10
)
y_values = np.linspace(
    df['team'].min(),
    df['team'].max(),
    10
)
xGrid, yGrid = np.meshgrid(x_values, y_values)

final = pd.DataFrame({
    'funding': xGrid.ravel(),
    'team': yGrid.ravel()
})

z_final = model.predict(final).reshape(10, 10)
fig = ex.scatter_3d(df, x='funding', y='team', z='revenue')
fig.add_trace(go.Surface(x=x_values, y=y_values, z=z_final, opacity=0.5, colorscale='Viridis'))
fig.show()
```
